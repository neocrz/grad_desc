# Gradiente Descendente

## para $f(x) = x^2+y^2$

```python
def grad(x,y):
    return(2*x,2*y) # gradiente composto pelas der.parciais
```

```python
def dist(x_n,y_n,x_o,y_o):
    return ((x_n-x_o)**2 + (y_n-y_o)**2)**0.5
```

```python
def grad_desc(x_o,y_o,tol, lr):
    x_n=x_o
    y_n=y_o
    x_n1=999999
    y_n1=999999
    i=0
    while True:
        x_n1=x_n-lr*grad(x_n,y_n)[0]
        y_n1=y_n-lr*grad(x_n,y_n)[1]
        i+=1
        if (dist(x_n1,y_n1,x_n,y_n)>tol):
            x_n=x_n1
            y_n=y_n1
        else:
            break
    return (x_n,y_n), i
```

```python
gd, i = grad_desc(10,10,10**(-16),0.01)
```
